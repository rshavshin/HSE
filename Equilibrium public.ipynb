# Price Competition in Finite Markets With Private Consumer Valuations

## The Model:
### 2 sellers with 1 unit of some indivisible good.
### n buyers with private valuations (the maximum price at which they are willing to buy the good).
### 1-st Stage: Sellers simultaneously and independently set prices for their goods.
### 2-nd Stage: Buyers simultaneously and independently decide to which seller, if anyone, to go.
### If more than one buyer come to the seller, the good will be sold  with equal probability.

## The equilibrium in the buyers subgame:
### If $v \in [0, p_l)~-$  go home;
### If $v \in [p_l, k)~-$  go to the seller with the lowest price;
### If $v \in [k, 1]~-$  go to the seller with the highest price.

```python
def make_data(n = 100):
    
    x = []
    
    for i in range(2, n + 1):
        x.append(equil(i))
    
    data = pd.DataFrame(x, columns = ['pl', 'ph', 'k', 'profit_pl', 'profit_ph', 'mean_U', 'expected_deals', 'Prob of successful deal for v = 1', 'expected_deals / number of buyers'])
    data.index += 2
    data['ph - pl'] = data['ph'] - data['pl']
    return data
```
